```python
N, T, H = 10, 5, 4

hs = np.random.randn(N, T, H)
hs
```

```python
print(hs.shape)
```

```python
a = np.random.randn(N, T)
a
```

```python
a
```

```python
ar = a.reshape(N, T, 1)
```

```python
ar
```

```python
print(ar.shape)
```

```python
import sys
sys.path.append('..')
import tensorflow as tf
import numpy as np
```

```python
N, T, H = 10, 5, 4
hs = np.random.randn(N, T, H)
hs
```

```python
h = np.random.randn(N, H)
h
```

```python
hr = h.reshape(N, 1, H).repeat(T, axis=1)
hr
```

```python
t = hs * hr
t
```

```python
s = np.sum(t, axis=2)
s
```

```python
sess = tf.Session()
soft = sess.run(tf.nn.softmax(s))
print(soft)
```
